E. Интерполяция
Ограничение времени	2 секунды
Ограничение памяти	64Mb
Ввод	стандартный ввод или input.txt
Вывод	стандартный вывод или output.txt
Есть некоторый многочлен второй степени от 5 переменных f(x1, x2, x3, x4, x5), с помощью которого были получены значения в некотором наборе точек. По этому набору точек и значениях в них надо предсказать значения в другом наборе точек с абсолютной или относительной точностью 10-6.

Формат ввода
На вход подаётся файл из 1000 тренировочных точек вместе со значением и 1000 тестовых точек.

Более формально сразу идёт 1000 строк тренировочной выборки, состоящих из 6 действительных чисел, разделённых табуляцией ('\t'). Первые 5 из этих 6 чисел – задают переменные . Последнее число задаёт значение функции в этой точке.

Затем идёт 1000 строк тестовой выборки, состоящих из 5 действительных чисел, разделённых табуляцией. Все они задают переменные 

Формат вывода
Нужно вывести 1000 строк, каждая из которой состоит из одного числа – значения функции в тестовой точки. Выводить значения в точках нужно в том же порядке, в котором точки тестовой выборки пришли во входном файле.

Примечания
Компилятор Python 3.7.3 ML содержит такие библиотеки, как sklearn, tensorflow, pytorch и несколько других.

```python
input_file = 'input.txt'
# input_file = '05_input01.txt'

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


with open(input_file, 'r') as f:
    X_train = []
    y_train = []

    X_test = []

    for i in range(1000):
        list_float = [float(x) for x in f.readline().split()]
        X_train.append(list_float[:5])
        y_train.append(list_float[-1])

    for i in range(1000):
        list_float = [float(x) for x in f.readline().split()]
        X_test.append(list_float[:5])        

    pipe = Pipeline([('scaller', StandardScaler()), ('poly', PolynomialFeatures(2)), ('linregr', LinearRegression())])

    pipe.fit(X_train, y_train)

    y_predict = pipe.predict(X_test)

    for i in y_predict:
        print(i)
```
