```python
import pandas as pd #прописываем используемые библиотеки
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
a = 3
gamma = sts.gamma(a) #задаём гамма-распределение с параметром a = 3
sample =  gamma.rvs(size=1000) #выборка объёма 1000
plt.hist(sample, normed = True,label='fraction of samples') #гистограмма
x = np.linspace(0,14, 100)
pdf = gamma.pdf(x)
plt.plot(x, pdf,label='theoretical') #теоретическая плотность распределения случайной величины
plt.legend()
plt.ylabel('f(x)')
plt.xlabel('x')
```

```python
import math
mean_np = np.zeros(1000) #
for n in [5,10,50]: 
    for count in range(1000):
        sample =  gamma.rvs(size = n) #для трёх значений n генерируем 1000 выборок
        mean_np[count] = sample.mean() #выписываем массив выборочных средних
    plt.figure(n) #на трёх графиках
    x = np.linspace(0,6,100)
    norm = sts.norm(3, math.sqrt(3.0/n)) #используем данные из теории о значении матожидания т дисперсии
    y = norm.pdf(x)
    plt.plot(x, y, label='theoretical') #рисуем график теоретической плотности нормального распределения  
    plt.hist(mean_np, normed = True, label='fraction of samples') #гистограммы распределений выборочных средних
    plt.legend()
    plt.title('n = %s' % n)
    plt.ylabel('f(x)')
    plt.xlabel('x')
    
    
    
```

## Выкладки с вычислениями параметров нормальных распределений
Для гамма-распределения с условиями $gamma=3$, $beta=1$ значения матожидания и дисперсии будут одинаковы и равны $EX = DX = 3$.
Тогда из центральной предельной теоремы следует, что выборочное среднее можно приблизить нормальным распределением $N(EX,DX/n)=N(3,3/n)$, где n = {5,10,50}

Как можно заметить с увеличением объёма выборок происходит сглаживание, и распределение принимает значение всё ближе к нормальному. То есть точность нормальной аппроксимации увеличивается. Данные графики наглядно демонстрирует работу центральной предельной теоремы.
Так же с ростом числа n гистограма сужается к значению $EX=3$.
