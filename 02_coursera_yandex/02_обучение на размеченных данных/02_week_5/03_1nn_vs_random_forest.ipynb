### Определение дополнительных функций

```python
def write_ans(file_name, ans):
    with open(file_name, 'w') as file:
        file.write(str(ans))
```

### Задание 1

```python
from sklearn import datasets
digits = datasets.load_digits(as_frame=True)

print(type(digits))
print(digits.keys())
print(digits.DESCR)
```

```python
digits.data
```

```python
from sklearn.metrics import accuracy_score
from knn import NeighborClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    digits.data.to_numpy(), digits.target.to_numpy(), test_size=0.25, shuffle=False, random_state=0)
```

```python
n_clf = NeighborClassifier(X_train, y_train)
y_pred = n_clf.predict(X_test)
from sklearn.metrics import accuracy_score

res1 = accuracy_score(y_true=y_test, y_pred=y_pred)
res1 = 1 - res1 # для определения доли ошибок
write_ans('03_1nn_ans_01', res1)
res1
```

### Задание 2

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=1000)

rf_clf.fit(X_train, y_train)
y_rf_pred = rf_clf.predict(X_test)
```

```python
res2 = 1 - accuracy_score(y_true=y_test, y_pred=y_rf_pred) # определяем долю ошибок

write_ans('03_1nn_ans_02', res2)
res2
```

Обратите внимание, что качество заметно хуже, чем у 1NN. Это нестандартная ситуация, как правило Random Forest отрабатывает лучше, но такие ситуации иногда встречаются.
