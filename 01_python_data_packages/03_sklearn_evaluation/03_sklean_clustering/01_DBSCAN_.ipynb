```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, 2], [2, 2], [2, 3], [8, 7], [8, 8], [25, 80]])
x_plot = []
y_plot = []

for dot in X:
    x_plot.append(dot[0])
    y_plot.append(dot[1])

plt.scatter(x_plot, y_plot)
```

```python
clustering = DBSCAN(eps=3, min_samples=2).fit(X)
clustering.labels_
plt.scatter(x_plot, y_plot, c=clustering.labels_)
```

```python
mx = [[i + j*i > 60 for i in range(1, 10)]  for j in range(1, 10)]
print(np.array(mx))
cors = []
vals = []
points = []

for ni, nj in enumerate(mx):
    for mi, mj in enumerate(nj):
        points.append(np.array([ni, mi, mj]))

plot_cors = list(zip(*points))
plt.scatter(plot_cors[0], plot_cors[1], c=plot_cors[2])
clustering = DBSCAN(eps= 1, min_samples=1).fit(points)
clustering.labels_
```

```python
x = list(zip(*clustering.components_))[0]
y = list(zip(*clustering.components_))[1]
c = list(zip(*clustering.components_))[2]

plt.scatter(x, y, c=c)
```

```python
import numpy as np

dict_ind = {}

for i in list(set(clustering.labels_)):
    print(i)
    dict_ind[i] = np.where(clustering.labels_ == i)
```

```python
dict_coors = {}

for key in dict_ind.keys():
    dict_coors[key] = np.take(clustering.components_[:, 0:2], dict_ind[key], 0)
```

```python
coors = dict_coors[1]
coors
```
