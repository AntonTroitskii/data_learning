```python
import numpy as np
from scipy.ndimage import shift as shift
import matplotlib.pyplot as plt

n = 10
mid_ind = n // 2

arr = np.zeros(n)
array_shifted_list = []
shift_=0.1

for i in range(mid_ind- 1, mid_ind+1):
    arr[i] = 1

print(arr)
for i in range (1, 6):
    array_shifted = np.zeros(arr.shape[0])
    shift(input=arr, shift=shift_, output=array_shifted, order=i)
    array_shifted_list.append([array_shifted, "order={}".format(i)])
    print(array_shifted)

plt.plot(arr)
for elem in array_shifted_list:
    plt.plot(elem[0], label=elem[1])

plt.legend()
plt.title('Dependence from order. Shift=0.1')
print('Можно видеть, что смещенине работает с некоторыми особенностями, связанными с интерполяцией трапецивидной функции.')
```

```python
import numpy as np
from scipy.ndimage import shift as shift
import matplotlib.pyplot as plt

n = 10
mid_ind = n // 2

arr = np.zeros(n)
array_shifted_list = []
shift_=1

for i in range(mid_ind- 1, mid_ind+1):
    arr[i] = 1

print(arr)
for i in range (1, 6):
    array_shifted = np.zeros(arr.shape[0])
    shift(input=arr, shift=shift_, output=array_shifted, order=i)
    array_shifted_list.append([array_shifted, "order={}".format(i)])
    print(array_shifted, "order={}".format(i))

plt.plot(arr)
for elem in array_shifted_list:
    plt.plot(elem[0], label=elem[1])

plt.legend()
plt.title('Dependence from order. Shift = 1')
print('Можно видеть, что интерполяция работает пркрасно при shift=1 для трапецивидной функции.')
print('Также можно заметить, что для интерполяции функциями 1 порядка, значения смещеннной фукнции просто сдвигаются в право.')
```
